Using screw image code the following cell:

```python
#read image
import cv2
import matplotlib.pyplot as plt
screw = cv2.imread("./data/screw.png")
```

```python
screw.shape
```

```python
plt.imshow(screw)
```

```python
screw_gray = cv2.cvtColor(screw, cv2.COLOR_BGR2GRAY)
ret, screw_bin = cv2.threshold(screw_gray, 250, 255, cv2.THRESH_BINARY)

screw_bin.shape
# plt.imshow(screw_bin, cmap="gray")
```

Q: draw & plot the contour (inner) --> Child of circle which is inside the right shape

```python
conts, hierarchy = cv2.findContours(screw_bin, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
print("Number of Contours:", len(conts))
```

```python
hierarchy
```

```python
screw_copy1 = screw.copy()
screw_copy2 = screw.copy()

# First Way
result = cv2.drawContours(screw_copy1, conts, 2, (255,0,0), 3)
plt.imshow(result)
```

```python
# Second Way
result = cv2.drawContours(screw_copy2, conts, hierarchy[0][1][2], (0,0,255), 3)
plt.imshow(result)
```
