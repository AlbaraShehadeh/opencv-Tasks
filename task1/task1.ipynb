<b>Task 1 : Image Transformation using OpenCV</b>

Using mr bean image code the following cells:

```python
#read image
import cv2
import numpy as np
import matplotlib.pyplot as plt
bean = cv2.imread("./data/bean.png")
```

Q1: Image Translation -> shift of an image from one location to another

```python
#start code


tx = 70  
ty = 50  
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])

# Apply the translation using warpAffine
bean_translated = cv2.warpAffine(bean, translation_matrix, (bean.shape[1], bean.shape[0]))

# Display the original and translated images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1), plt.imshow(cv2.cvtColor(bean, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(1, 2, 2), plt.imshow(cv2.cvtColor(bean_translated, cv2.COLOR_BGR2RGB)), plt.title('Translated Image')
plt.show()

#end code
```

Q2: Image Reflection -> flip the image horizontally & vertically

```python
#start code

bean_horizontal = cv2.flip(bean, 1) # horizontally
bean_vertical = cv2.flip(bean, 0) # vertically


plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1), plt.imshow(cv2.cvtColor(bean, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(1, 3, 2), plt.imshow(cv2.cvtColor(bean_horizontal, cv2.COLOR_BGR2RGB)), plt.title('Horizontal Flip')
plt.subplot(1, 3, 3), plt.imshow(cv2.cvtColor(bean_vertical, cv2.COLOR_BGR2RGB)), plt.title('Vertical Flip')

plt.show()

#end code
```

Q3: Image Rotation -> rotate the image in any degree you want eg: (30, 60, 90)

```python
#start code

angles= [30, 60, 90]
bean_rotated = []

for angle in angles:
    bean_rotated.append(cv2.warpAffine(bean, cv2.getRotationMatrix2D((bean.shape[1] // 2, bean.shape[0] // 2), angle, 1.0),
                          (bean.shape[1], bean.shape[0]))
)

# Display the original and rotated images
plt.figure(figsize=(12, 4))
plt.subplot(1, len(angles) + 1, 1), plt.imshow(cv2.cvtColor(bean, cv2.COLOR_BGR2RGB)), plt.title('Original Image')

for i, angle in enumerate(angles):
    plt.subplot(1, len(angles) + 1, i + 2), plt.imshow(cv2.cvtColor(bean_rotated[i], cv2.COLOR_BGR2RGB)), plt.title(f'Rotated {angle} degrees')

plt.show()

#end code
```

Q4: Image Scaling -> resize image in any height and width you want eg: (480,640), (720, 800)

```python
#start code

bean_resized_1 = cv2.resize(bean, (640, 480)) # (480, 640)
bean_resized_2 = cv2.resize(bean, (800, 720)) # (720, 800)

# Display the original and resized images
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1), plt.imshow(cv2.cvtColor(bean, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(1, 3, 2), plt.imshow(cv2.cvtColor(bean_resized_1, cv2.COLOR_BGR2RGB)), plt.title('(480, 640)')
plt.subplot(1, 3, 3), plt.imshow(cv2.cvtColor(bean_resized_2, cv2.COLOR_BGR2RGB)), plt.title('(720, 800)')

plt.show()

#end code
```

Q5: Image Cropping -> remove the background pixel of image (not importent)

```python
#start code

print("Image Shpe (before): ", bean.shape)
bean_gray = cv2.cvtColor(bean, cv2.COLOR_BGR2GRAY)
print("Image Shpe (after): ", bean_gray.shape)
plt.imshow(bean_gray, cmap='grey')

#end code
```

```python
hsv_image = cv2.cvtColor(bean, cv2.COLOR_BGR2HSV)

lower_bg = np.array([0, 0, 128])
upper_bg = np.array([180, 50, 255])

background_mask = cv2.inRange(hsv_image, lower_bg, upper_bg) # binary mask
foreground_mask = cv2.bitwise_not(background_mask)
bean_cropped = cv2.bitwise_and(bean, bean, mask=foreground_mask)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1), plt.imshow(cv2.cvtColor(bean, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(1, 2, 2), plt.imshow(cv2.cvtColor(bean_cropped, cv2.COLOR_BGR2RGB)), plt.title('Cropped Image')
plt.show()


# This is best result i got
```

Q6: Image Shearing -> shear the image once on X-Axis & Y-Axis

```python
#start code

# X-Axis
bean_sheared_x = cv2.warpAffine(bean, np.float32([[1, 0.3, 0], [0, 1, 0]]), (bean.shape[1], bean.shape[0]))

# Y-Axis
bean_sheared_y = cv2.warpAffine(bean, np.float32([[1, 0, 0], [0.3, 1, 0]]), (bean.shape[1], bean.shape[0]))

# Display the original and sheared images
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1), plt.imshow(cv2.cvtColor(bean, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(1, 3, 2), plt.imshow(cv2.cvtColor(bean_sheared_x, cv2.COLOR_BGR2RGB)), plt.title('Sheared X-Axis')
plt.subplot(1, 3, 3), plt.imshow(cv2.cvtColor(bean_sheared_y, cv2.COLOR_BGR2RGB)), plt.title('Sheared Y-Axis')

plt.show()

#end code
```
